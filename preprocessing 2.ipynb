```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    try:
        result = tmdb_movie.search(x)
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            genre_list = str(genres)
            return genre_list
    except:
        genre_list = [np.NaN]
        return genre_list 
#         return genre_str.join(genre_list)
    else:
        np.NaN
```

```python
import pandas as pd
import numpy as np
```

```python
movies_data = pd.read_csv('movies_data_wikipedia.csv')
```

```python
movies_data
```

```python
## Convert genre list to string
def list_to_string(x):
    if x:
        try:
            return x.replace(',', '').replace('[','').replace(']','').replace("'",'')
        except:
            return np.NaN
    else:
        return np.NaN
```

```python
movies_data['genres'] = movies_data['Genre'].map(lambda x: list_to_string(x))
```

```python
movies_data
```

```python
# get director
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]


# get actor1 
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])

# get actor2
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])

#get actor3
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
movies_data['director_name'] = movies_data['Cast and crew'].map(lambda x: get_director(str(x)))
movies_data['actor_1_name'] = movies_data['Cast and crew'].map(lambda x: get_actor1(str(x)))
movies_data['actor_2_name'] = movies_data['Cast and crew'].map(lambda x: get_actor2(str(x)))
movies_data['actor_3_name'] = movies_data['Cast and crew'].map(lambda x: get_actor3(str(x)))
```

```python
movies_data
```

```python
movies_data = movies_data.rename(columns={'Title':'movie_title'})
```

```python
movies_data['comb'] = movies_data['actor_1_name'] + ' ' + movies_data['actor_2_name'] + ' '+ movies_data['actor_3_name'] + ' '+ movies_data['director_name'] +' ' + movies_data['genres']
```

```python
movies_data
```

```python
movies_data['actor_2_name'] = movies_data['actor_2_name'].replace(np.nan, 'unknown')
movies_data['actor_3_name'] = movies_data['actor_3_name'].replace(np.nan, 'unknown')
movies_data['director_name'] = movies_data['director_name'].replace(np.nan, 'unknown')
movies_data['movie_title'] = movies_data['movie_title'].replace(np.nan, 'unknown')
movies_data['actor_1_name'] = movies_data['actor_1_name'].replace(np.nan, 'unknown')
movies_data['comb'] = movies_data['comb'].replace(np.nan, 'unknown')
movies_data['genres'] = movies_data['genres'].replace(np.nan, 'unknown')
```

```python
movies_data['movie_title'] = movies_data['movie_title'].str.lower()
```

```python
movies_data = movies_data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title','comb']]
```
